```python
top2spec = pd.read_csv('top2spec.csv')
```

```python
rem = list(top2spec[top2spec['specialty'] == ' '].index)
```

```python
slider = widgets.IntSlider( value=0, min=0, max=len(rem), step=1, description='Index:', disabled=False, continuous_update=False,
                            orientation='horizontal', readout=True, readout_format='d')

lab = widgets.Label(value=top2spec.loc[rem[slider.value],'topic'], description='Topic:', disabled=False)
sel = widgets.SelectMultiple(options=spec, rows=10, description='Specialty', disabled=False)
btn = widgets.Button(description="Next")

def on_button_clicked(btn):
    global slider 
    global lab
    global sel
    global rem
    
    top2spec.loc[rem[slider.value], 'specialty'] = sel.value 
    top2spec.to_csv('top2spec.csv', index=False)

    slider.value = slider.value + 1
    lab.value = top2spec.loc[rem[slider.value],'topic']


btn.on_click(on_button_clicked)

display(slider)
display(lab)
display(sel)
display(btn)
```
